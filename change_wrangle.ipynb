```python
import pandas as pd

file = pd.read_csv('total_counties_02.csv')
counties_df = pd.DataFrame(file)
```

```python
#The original dataframe had a column 'Unnamed: 0' inserted. Reading the CSV to remove this 
#messed up my code below, so I figured it didn't matter too much for the size of what we're working with

counties_df.drop('Unnamed: 0', axis=1)
```

```python
#This cell gives us adds a column for yearly change into the dataframe (again, 'Unnamed: 0' showed up again)

yearly_change = []

try:
    for x in counties_df.index:
        change = counties_df.iloc[x+1, 4] - counties_df.iloc[x, 4]
        yearly_change.append(change)
except IndexError:
    pass

yearly_change.insert(0,0)

counties_df['yearly_change'] = yearly_change

counties_df.drop('Unnamed: 0', axis=1)
```

```python
#If y'all were wanting to examine the change between 2 years, the following function should do that.

def LE_change(year1,year2):
    df1 = counties_df.loc[counties_df['year_id'] == year1, :]
    df2 = counties_df.loc[counties_df['year_id'] == year2, :]
    change_list = []
    try:
        for x in df1.index:
            change = df2.iloc[x, 4] - df1.iloc[x, 4]
            change_list.append(change)
    except IndexError:
        pass
    return change_list

#For example
LE_change(1980,2014)
```

```python
counties_df.drop('Unnamed: 0', axis=1).to_csv('yearly_change', index=False)
```
