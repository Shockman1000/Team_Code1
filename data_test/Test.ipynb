```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = ('test_files/Alabama.CSV')

alabama = pd.read_csv(path)
alabama_df = pd.DataFrame(alabama)
```

```python
#Narrowing down which data we want to look at
alabama_df = alabama_df[((alabama_df.measure_name == 'Life expectancy') &   #Removes mortality rate
                        (alabama_df.sex == 'Both') &                        #Removes Male/Female; simplifies data
                        (alabama_df.location_name != 'Alabama') &           #Removes data for state as a whole
                        (alabama_df.year_id >= 2010))]                      #Limit years to 2010-2014 (arbitrary)  

alabama_df
```

```python
#remove unneeded columns
alabama_df = alabama_df.drop(['measure_id', 'location_id', 'FIPS', 'sex_id', 'age_id', 'metric', 'upper', 'lower'],axis=1)
```
